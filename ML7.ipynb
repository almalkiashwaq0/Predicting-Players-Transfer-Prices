```python
df.drop(['name' , 'position','goals'], axis=1 , inplace=True)
```

```python
categorical_features = ['team']
numeric_features = ['height' 'age',	'appearance','assists','yellow cards','second yellow cards','red cards','goals conceded','clean sheets','minutes played','days_injured','games_injured',
                    'award','current_value','highest_value','position_encoded',	'winger']
```

```python
# one hot coding
df = pd.get_dummies(df, columns=categorical_features)
```

```python
correlation = df.corr()
print(correlation['current_value'].sort_values(ascending=False))
```

```python
threshold = 0.2  # You can change this value based on your requirement

selected_features = correlation[abs(correlation['current_value']) > threshold]['current_value'].index
selected_features
```

```python
selected_features = ['appearance', 'minutes played', 'current_value', 'highest_value']
```

```python
df = df[selected_features]
df.head()
```

```python
# Prepare data
X = df.drop(['current_value'], axis=1)
y = df['current_value']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    shuffle=True,
                                                    random_state=42)

# sacle the data
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LinearRegression()
```

```python
# Fit the model on the training data
model.fit(X_train_scaled, y_train)
```

```python
# Predict and evaluate the model
y_pred = model.predict(X_test_scaled)
```

```python
coeff_df = pd.DataFrame(model.coef_,X.columns,columns=['Coefficient'])
coeff_df
```
